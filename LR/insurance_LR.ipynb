```python
for i in col:
    if df[i].dtype == "object":
        df[i]=df[i].fillna(df[i].mode())

    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

```python
#duplicate values
df.duplicated().sum()



#if duplicate is present
#use drop.duplicates
```

```python
#outliers
for i in col:
    if df[i].dtype != "object":
        plt.boxplot(df[i])
        plt.title(i)
        plt.show()
```

```python
for i in col:
    if df[i].dtype !="object":
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        
        iqr=q3-q1

        lower_bond= q1-1.5*iqr
        upper_bond= q3+1.5*iqr

        df = df[(df[i] >= lower_bond) & (df[i] <= upper_bond)]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns= ["charges"])  #all indendent columns
y=df["charges"]   #target columns
```

```python
df
```

* train ---> Traning
* test ---> Testing

* Learning Phase
* x_train ---> Traning Phase
* y_train ---> Traning Aswers


* Testing Phase
* x_test ---> Testing The Qustions
* y_pred ---> Model Answers/Mechine Answers

* Compare:
* y_pred VS y_test

### **Term**	  |   **Meaning**
### **x_train**	  |  questions used to teach student
### **y_train**	  |   correct answers
### **x_test**	  |   new questions in exam
### **y_test**	  |   real exam answers
### **y_pred**	  |   student's predicted answers

```python
x_tarin,x_test,y_train,y_test= train_test_split(x,y,train_size=0.70,random_state=42)
```

```python
my_model=LinearRegression()
```

```python
#traning
my_model.fit(x_tarin, y_train)
```

```python
#tasting
y_pred= my_model.predict(x_test)
```

```python
pd.DataFrame(y_pred)
```

```python
result= pd.DataFrame(columns=["Actual_values","Preducted_values"])
result["Actual_values"]=y_test
result["Preducted_values"]=y_pred
```

```python
result
```

```python
from sklearn.metrics import * 
```

```python
result =r2_score(y_test,y_pred)
result
```

```python
mes= mean_absolute_error(y_test,y_pred)
mes
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
#rmse
rmse =np.sqrt(mse)
rmse
```
